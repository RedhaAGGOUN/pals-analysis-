# ðŸ“Š Analyse Exploratoire des Pals (Palworld Dataset)
Ce notebook rÃ©pond aux questions a â†’ x de l'analyse exploratoire Ã  partir des fichiers nettoyÃ©s.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“ Chargement des fichiers nettoyÃ©s
combat_df = pd.read_csv('../2_cleaned/combat_attribute_clean.csv')
job_df = pd.read_csv('../2_cleaned/job_skill_clean.csv')

# ðŸ”¤ Normalisation des noms de colonnes
combat_df.columns = combat_df.columns.str.strip().str.lower().str.replace(' ', '_')
job_df.columns = job_df.columns.str.strip().str.lower().str.replace(' ', '_')

# ðŸ”— Fusion des datasets sur le nom
combat_df['name'] = combat_df['name'].str.strip().str.lower()
job_df['english_name'] = job_df['english_name'].str.strip().str.lower()
merged_df = pd.merge(combat_df, job_df, left_on='name', right_on='english_name', how='inner')
```

## a. Distribution de la taille des Pals (`volume_size`)

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=merged_df, x='volume_size_x', order=merged_df['volume_size_x'].value_counts().index)
plt.title('a. Distribution de la taille des Pals')
plt.xlabel('Taille (volume_size)')
plt.ylabel('Nombre de Pals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## b. Distribution des catÃ©gories (`genuscategory`)

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=merged_df, x='genuscategory', order=merged_df['genuscategory'].value_counts().index)
plt.title('b. Distribution des catÃ©gories des Pals (genuscategory)')
plt.xlabel('CatÃ©gorie biologique')
plt.ylabel('Nombre de Pals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## c. Distribution des points de vie (HP)

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=merged_df, x='hp', bins=20, kde=True)
plt.title('c. Distribution des points de vie (HP)')
plt.xlabel('Points de vie (HP)')
plt.ylabel('Nombre de Pals')
plt.tight_layout()
plt.show()
```

## d. Distribution de la raretÃ© (rarity)

```python
# ðŸŽ¯ Filtrage des raretÃ©s valides (ex. de 1 Ã  10)
valid_rarity_df = merged_df[(merged_df['rarity'] >= 1) & (merged_df['rarity'] <= 10)]

# ðŸ“Š Graphique de distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=valid_rarity_df, x='rarity', order=sorted(valid_rarity_df['rarity'].unique()))
plt.title("d. Distribution de la raretÃ© des Pals (valeurs 1 Ã  10 uniquement)")
plt.xlabel("RaretÃ©")
plt.ylabel("Nombre de Pals")
plt.tight_layout()
plt.show()
```

## e. Distribution de la consommation alimentaire (food_intake)

```python
# Filtrage de la plage valide
valid_food_df = merged_df[merged_df['food_intake'].between(1, 6)]

# Re-gÃ©nÃ©ration du graphique
plt.figure(figsize=(10, 5))
sns.countplot(data=valid_food_df, x='food_intake', order=sorted(valid_food_df['food_intake'].unique()))
plt.title("e. Distribution de la consommation alimentaire des Pals (filtrÃ©e : 1 Ã  6)")
plt.xlabel("Consommation alimentaire")
plt.ylabel("Nombre de Pals")
plt.tight_layout()
plt.show()
```

## f. Pals utiles Ã  la production (ranch : laine, Å“uf, laitâ€¦)

```python
ranch_items = merged_df[merged_df['ranch_items'].notna()][['name', 'ranch_items']].sort_values('ranch_items')
print('f. Pals utiles Ã  la production (ranch items) :')
print(ranch_items)
```

## g. Top 10 des Pals les plus puissants (par attaque et dÃ©fense combinÃ©es)

```python
merged_df['total_power'] = merged_df[['melee_attack', 'remote_attack', 'defense']].sum(axis=1, numeric_only=True)
top_powerful = merged_df[['name', 'melee_attack', 'remote_attack', 'defense', 'total_power']].sort_values(by='total_power', ascending=False).head(10)
print('g. Top 10 des Pals les plus puissants :')
print(top_powerful)
```

## h. CorrÃ©lations entre les attributs de combat

```python
combat_stats = merged_df[['hp', 'melee_attack', 'remote_attack', 'defense', 'support', 'speed_of_work']].dropna()
plt.figure(figsize=(10, 6))
sns.heatmap(combat_stats.corr(), annot=True, cmap="coolwarm")
plt.title("h. CorrÃ©lation entre les attributs de combat")
plt.tight_layout()
plt.show()
```

## i. Impact de la raretÃ© sur les attributs de base (HP, attaque, dÃ©fense)

```python
# âœ… Filtrer uniquement les raretÃ©s valides (1 Ã  10)
valid_rarity_df = merged_df[merged_df['rarity'].between(1, 10)]

# ðŸ“Š Graphique corrigÃ©
plt.figure(figsize=(10, 6))
sns.boxplot(data=valid_rarity_df, x='rarity', y='hp')
plt.title("i. HP en fonction de la raretÃ© (filtrÃ©e entre 1 et 10)")
plt.xlabel("RaretÃ©")
plt.ylabel("HP")
plt.tight_layout()
plt.show()
```

## j. RaretÃ© moyenne des Pals Ã  forte attaque

```python
strong_pals = merged_df[merged_df['melee_attack'] >= merged_df['melee_attack'].quantile(0.9)]
moyenne_raretÃ© = strong_pals['rarity'].mean()
print(f"RaretÃ© moyenne des Pals avec attaque Ã©levÃ©e (top 10%) : {moyenne_raretÃ©:.2f}")
```

## k. Influence de la taille sur la performance au combat

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df, x='volume_size_x', y='total_power')
plt.title("k. Influence de la taille sur la puissance totale")
plt.xlabel("Taille")
plt.ylabel("Puissance totale (attaque + dÃ©fense)")
plt.tight_layout()
plt.show()
```

## l. Les Pals rapides sont-ils plus efficaces ?

```python
# ðŸ’¡ Calcul de la puissance totale
merged_df['total_power'] = merged_df[['melee_attack', 'remote_attack', 'defense']].sum(axis=1, numeric_only=True)

# ðŸ“Š Nuage de points avec ligne de rÃ©gression
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='running_speed', y='total_power', label='Pals')
sns.regplot(data=merged_df, x='running_speed', y='total_power', scatter=False, color='red', label='Tendance')
plt.title("l. Vitesse vs. Puissance totale (avec ligne de rÃ©gression)")
plt.xlabel("Vitesse de course")
plt.ylabel("Puissance totale")
plt.legend()
plt.tight_layout()
plt.show()
```

## m. Proposer une Ã©quipe Ã©quilibrÃ©e (5 Pals aux attributs complÃ©mentaires)

```python
balanced_team = merged_df.sort_values(by=['total_power', 'speed_of_work'], ascending=False).drop_duplicates(subset=['genuscategory']).head(5)
print('m. Ã‰quipe Ã©quilibrÃ©e de 5 Pals :')
print(balanced_team[['name', 'genuscategory', 'hp', 'melee_attack', 'defense', 'speed_of_work', 'total_power']])
```

## s. Pals avec la vitesse de travail la plus Ã©levÃ©e

```python
fast_workers = merged_df[['name', 'handling_speed']].sort_values(by='handling_speed', ascending=False).dropna().head(10)
print('s. Pals avec la vitesse de travail la plus Ã©levÃ©e :')
print(fast_workers)
```

## u. Boss avec attributs de combat les plus forts (total des stats)

```python
boss_df = pd.read_csv('../2_cleaned/ordinary_boss_clean.csv')
boss_df.columns = boss_df.columns.str.strip().str.lower().str.replace(' ', '_')
boss_df['combat_score'] = boss_df[['hp', 'remote_attack']].sum(axis=1, numeric_only=True)
strongest_boss = boss_df[['name', 'hp', 'remote_attack', 'combat_score']].sort_values(by='combat_score', ascending=False).head(1)
print('u. Boss avec le score de combat le plus Ã©levÃ© :')
print(strongest_boss)
```
